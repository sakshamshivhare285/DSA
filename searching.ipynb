```python
# Linear Search

def linearSearch(arr,x):
  for i in range(len(arr)):
    if arr[i] == x:
      return i
  return -1

# time complexcity = O(n)
# space complexcity = O(1)

# driver code
arr = [20,45,27,47,55,67,75,88,90]

result = linearSearch(arr,67)
print('searched element is prsent at ,',result)
```

```python
def binarySearch(arr,left,right,x):
  while(left<=right):
      mid = int(left + ((right-left)//2))
      if arr[mid] == x:
        return ('value you are looking for is at', mid)
      elif arr[mid] < x:
        # moves towards right side of mid
        # hence adding mid +1 to mid
        left = mid + 1
        return binarySearch(arr,left,right,x)
      else:
        # move towards lower side
        # right index becomes mid - 1
        right = mid-1
        return binarySearch(arr,left,right,x)
  return -1

# driver code
arr =[10,20,30,40,50,60,70,80,90]
x = 80
result = binarySearch(arr,0,8,100)
result
```

```python
def binarySearch_n(arr,left,right,x):
  while(left<=right):
    mid = int(left + (right - left)//2)
    if arr[mid] == x:
      return mid
    elif arr[mid] < x:
      left = mid +1
    else:
      right = mid -1
  return -1

# driver code
arr =[10,20,30,40,50,60,70,80,90]
x = 80
result = binarySearch_n(arr,0,0,10)
result
```
