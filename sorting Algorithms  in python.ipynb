**Implementation of bubble sort in python**:

---
Time complexcity is O(n^2)


```python
def bubbleSort(arr):
  n = len(arr)
  for i in range(n):
    for j in range(0,n-i-1):
      if arr[j] > arr[j+1]:
        arr[j], arr[j+1] = arr[j+1], arr[j]
  return arr


# driver code
arr = [70,20,50,30,90,5,15]
result = bubbleSort(arr)
print(result)

arr1 = [0,5,56,34,656,34,554,4,2]
results = bubbleSort(arr1)
print(results)
```

Selection sort in python

```python
def selectionsort(arr):
  n = len(arr)
  for i in range(n):
    min_idx = i
    for j in range(i+1,n):
      if arr[j] < arr[min_idx]:
        min_idx = j
    arr[i],arr[min_idx] = arr[min_idx], arr[i]
  return arr


# driver code
arr = [70,20,50,30,90,5,15]
result = selectionsort(arr)
print(result)
arr1 = [0,5,56,34,656,34,554,4,2]
results = selectionsort(arr1)
print(results)
```

 Insertion sort implementation

```python
# time complexcity is O(n^2)

def insertionSort(arr):
  n = len(arr)
  for i in range(1,n):
    key = arr[i]
    j = i - 1
    while(j>=0 and key< arr[j]):
      arr[j+1] = arr[j]
      j = j - 1
    arr[j+1] = key
  return arr

## driver code
arr = [75,90,95,100,85,80]
result = insertionSort(arr)
print(result)


```
