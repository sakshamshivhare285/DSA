Best time to sell and buy stocks problem leetcode

Rules for initializing
*  use bigger values for minprice, min value
*  Use zero for all max values



```python
def maxProfit(price):
  minprice = float('inf')
  maxprofit = 0

  for i in range(len(price)):
    if price[i] < minprice:
      minprice = price[i]
    elif price[i] - minprice > maxprofit:
      maxprofit = price[i] - minprice
  return maxprofit

# driver code
price = [7,1,5,3,6,24]
maxProfit(price)
```

2D array question

```python
def findingtarget(arr,target):
  # number of rows
  m = len(arr)
  if m == 0:
    return False

  # number of columns
  n = len(arr[0])
  left , right = 0 , m*n-1

  # using binary search
  # row_number ---> mid//n
  # column_number ---> mid%2
  while left<=right:
    mid = left + (right-left)// 2
    mid_element = arr[mid//n][mid%n]

    if mid_element == target:
      return True
    elif mid_element> target:
      right  = mid -1
    else:
      left = mid + 1
  return False


# driver code
arr = [[1,3,5,7],[10,11,16,20],[23,30,34,60]]
findingtarget(arr,60)
```
