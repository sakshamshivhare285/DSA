```python
class LinkedList:

  def __init__(self):

    # Empty Linked List
    self.head = None
    # no of nodes in the LL
    self.n = 0

  def __len__(self):
    return self.n

  def insert_head(self,value):

    # new node
    new_node = Node(value)

    # create connection
    new_node.next = self.head

    # reassign head
    self.head = new_node

    # increment n
    self.n = self.n + 1
  def __str__(self):
    curr = self.head
    result = ''
    while curr != None:
      result = result + str(curr.data) + '-->'
      curr = curr.next
    return result[:-3]

  # adding append function
  def append(self, value):

    next_node = Node(value) # creating a node from a vale

    if self.head == None:
      self.head = next_node
      self.n = self.n + 1
      return

    else:
    # tranvsering through the already created ll to get at the last position
      curr = self.head
      while curr.next != None:
        curr = curr.next
      # when the loop ends you reach at last location
      curr.next = next_node
      self.n = self.n + 1

    # item insert at a location
  def insert_after(self,after,value):
    new_node = Node(value)

    curr = self.head
    while curr != None:
      if curr.data == after:
        break
      curr =  curr.next

    if curr != None:
      new_node.next = curr.next
      curr.next = new_node
      self.n +=1
    else:
      'item not found'
# clear function
  def clear(self):
    self.head = None
    self.n = 0

  def clear_head(self):
    if self.head == None:
      return 'empty list'
    self.head = self.head.next
    self.n = self.n -1

# pop
  def pop(self):
    if self.head == None:
      return 'empty LL'
    curr = self.head
    if curr.next == None:
      return self.clear_head()

    while curr.next.next != None:
      curr = curr.next
    # current is second last item or node
    curr.next = None
    self.n = self.n-1
# remove
  def remove(self,value):

    if self.head == None:
      return 'empty ll'

    if self.head.data == value:
      # to remove head node
      return self.clear_head()
    curr = self.head
    while curr.next != None:
      if curr.next.data == value:
        break
      curr = curr.next

      if curr.next == None:
        # item not find
        return 'not found'
      else:
        curr.next = curr.next.next

# adding search feature
  def search(self,item):
    curr = self.head
    pos = 0

    while curr != None:
      if curr.data == item:
        return pos

      curr = curr.next
      pos = pos+1

    return 'Not found'

# addind indexing
  def __getitem__(self,index):
    curr = self.head
    pos = 0
    while curr != None:
      if pos == index:
        return curr.data
      curr = curr.next
      pos = pos + 1
    return "value not found"

# adding max value
  def replace_max(self, value):
    temp = self.head
    max = temp

    while temp!= None:
      if temp.data > max.data:
        max = temp
      temp = temp.next

    max.data =  value

## finding sum of odd index element
  def sum_odd(self):

    temp = self.head
    pos = 0
    result = 0

    while temp != None:
      if (pos % 2) != 0 :
        result = result + temp.data
      pos = pos + 1
      temp = temp.next

    return result
```
