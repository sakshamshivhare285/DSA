# Creating a class for linked List

```python
class Node:
  def __init__(self,data):
    self.data = data
    self.nextpointer = None

class LinkedList:
  def __init__(self):
    self.head = None
# inserting new node at the start of the linked List
  def insert_at_start(self,new_data):
    new_node = Node(new_data)
    # now insertion at the start
    new_node.nextpointer = self.head
    self.head = new_node
# printing linkedlist
  def printlinkedlist(self):
    temp = self.head
    while temp:
      print(str(temp.data) + " ", end=" ")
      temp = temp.nextpointer

  # inserting data at the end
  def insertAtEnd(self,new_data):
    # creating new node
    new_node = Node(new_data)
    if self.head == None:
      self.head = new_node
    else:
      temp = self.head
      while temp.nextpointer is not None:
        temp = temp.nextpointer
      temp.nextpointer = new_node

# inserting element after a given element
  def insertAfterNode(self,previous_node,new_data):
    if previous_node is None:
      return ('Value error: Given node not present in the Linked List')
    # create a new node
    new_node = Node(new_data)
    new_node.nextpointer = previous_node.nextpointer
    previous_node.nextpointer = new_node

  def insertNodeAfterValue(self,pre_node,new_data):
    new_node = Node(new_data)
    if self.head == None:
      self.head = new_node
    else:
      temp = self.head
      while temp.data != pre_node:
        temp = temp.nextpointer
      pre_node_pointer = temp.nextpointer
      next_node_pointer = temp.nextpointer
      new_node.nextpointer = next_node_pointer
      temp.nextpointer = new_node
```
