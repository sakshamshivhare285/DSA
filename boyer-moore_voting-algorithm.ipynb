Implementation of Biyer-moore algorithm to find majority element

```python

def findcandidate(arr):
  candidate = None
  count = 0
  for i in arr:
    if count == 0:
      candidate = i
    count += (1 if i == candidate else -1 )

  return candidate

#ismajority
def ismajorityElement(arr,candidate):
  count = 0
  for i in range(len(arr)):
    if arr[i] == candidate:
      count +=1
  if count < len(arr)/2:
    return 1
  else:
    return -1

# method defination
def voting(arr):
  candidate = findcandidate(arr)
  if ismajorityElement(arr,candidate):
    print('majority elements is',candidate)
  else:
    print('No element is there')



#driver code
nums = [2,2,1,1,1,2,2]
voting(nums)
```
