Implementation of merge sort using divide and conquer algorithm

```python


def MergeProcedure(arr,i,mid,j):
  # number of elements in left sub tree
  n1 = mid - i +1
  # number of elements in right sub tree
  n2 = j - mid
  # initializing left and right sub array
  LeftSubarray = [0] * n1
  RightSubarray = [0] * n2
  # adding values to left and right sub array
  for m in range(n1):
    LeftSubarray[m] = arr[i + m]

  for n in range(n2):
    RightSubarray[n] = arr[(mid +1) + n]

  p = 0
  q = 0
  k = i

  while p<n1 and q<n2:
    if LeftSubarray[p] < RightSubarray[q]:
      arr[k] = LeftSubarray[p]
      p += 1
    else:
      arr[k] = RightSubarray[q]
      q += 1
    k +=1
  # coping entire elemets from left array to arr
  while p < n1:
    arr[k] = LeftSubarray[p]
    p += 1
    k += 1
  # coping entire elemets from left array to arr
  while q < n2:
    arr[k] = RightSubarray[q]
    q += 1
    k +=1


def MergeSort(arr,i,j):
  if i < j:
    # divide the array
    mid = i +(j-i)//2
    ## conquer
    # recurcive call left sub tree
    MergeSort(arr,i,mid)
    #recurcive call right sub tree
    MergeSort(arr,mid+1,j)

    # combine using merge procedure
    MergeProcedure(arr,i,mid,j)
  return arr

# driver code
arr = [50,20,40,70,10,60,39,64]
i = 0
j = len(arr) -1
result  = MergeSort(arr,i,j)
print(result)
```
