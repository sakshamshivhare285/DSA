```python
# quick sort
# method defination for partition function
def Partition_function(arr,p,q):
  pivot_element = arr[p]
  i = p

  for j in range(i+1,q+1):
    if arr[j] <= pivot_element:
      i += 1
      # swap between value of arr[i] & arr[j]
      arr[i],arr[j] = arr[j], arr[i]
    # once out of loop final swap arr[i] = pivot_elemey
  arr[i],arr[p] = arr[p], arr[i]
  return i


def Quick_sort(arr,p,q):


  if p<q:
    ## use divide and conquer method
    # divide
    # calling partition function
    mid = Partition_function(arr,p,q)

    # call recurcive function call for left subtree
    Quick_sort(arr,p,mid-1)
    Quick_sort(arr,mid+1,q)

  return arr






# driver code
arr = [50,70,29,67,12,15,46]
p = 0
q = len(arr)-1
print(Quick_sort(arr,p,q))
```
