```python
# fuction for ternary search
#time complexcity O(log n base 3)

def ternarysearch(arr,left,right,key):

  mid1 =  left + (right-left)//3
  mid2 = right - (right-left)//3
  while left <right:

    if arr[mid1] == key:
      return mid1
    elif arr[mid2] == key:
      return mid2

    elif arr[mid1] > key:
      mid1 = mid1 -1
      return ternarysearch(arr,left,mid1,key) #keeo chaning the value for left and right depending upon the range

    elif arr[mid2] < key:
      mid2 = mid2 + 1
      return ternarysearch(arr,mid2, right,key)#keeo chaning the value for left and right depending upon the range
# if key lies between m1 and m2 range
    else:
      mid1 = mid1+1
      return ternarysearch(arr,mid1, mid2,key)#keeo chaning the value for left and right depending upon the range
  return -1

# driver code
arr = [20,25,47,56,59,63,65,79,82]
n = len(arr)
ternarysearch(arr,0,n,20)
```
