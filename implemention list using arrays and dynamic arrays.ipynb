```python
import ctypes
```

```python
class Meralist:
  def __init__(self):
    self.size = 1
    self.n = 0
    # create C type array with size as input
    self.A = self.__make_array(self.size)

  def __make_array(self,capacity):
    # create a C type array(static, dynamic) with size, capacity
    return (capacity*ctypes.py_object)()
  def __len__(self):
    return self.n
  def __resize(self,new_capacity):
    # create a new array with new capaciyt
    B = self.__make_array(new_capacity)
    self.size = new_capacity
    #copy content of A to B
    for i in range(self.n):
      B[i] = self.A[i]
    # reassign A
    self.A = B
  # adding append functionality
  def append(self, item):
    if self.n == self.size:
      # need to resize
      self.__resize(self.size*2)
    # append
    self.A[self.n] = item
    self.n = self.n + 1
# for printing
  def __str__(self):
    result = ''
    for i in range(self.n):
      result = result + str(self.A[i]) + ','
    return '[' + result[:-1] +']'


## adding indexing same as list using magic methof
  def __getitem__(self, index):
    if 0<= index < self.n:
      return self.A[index]
    else:
      return 'IndexError-->>> index out of range'

## adding pop function
  def pop(self):
    if self.n == 0:
      print('[]')
    else:
      print(self.A[self.n - 1])
      self.n = self.n - 1
## adding clear function to empty the list
  def clear(self):
    # we bring the array to initail state as when it was created
    self.n = 0
    self.size = 1

# for finding the index of passed element
  def find(self,item):
    for i in range(self.n):
      if self.A[i] == item:
        return i
    return ('value error---->>> value not in list')

# insert operation
  def insert(self,pos,item):
    # check in there empty position for the new elenemet
    if self.n == self.size:
      self.__resize(self.size*2)

    for i in range(self.n,pos,-1):
      self.A[i] = self.A[i-1]
    self.A[pos] = item
    self.n = self.n +1
# implementing delete function
  def __delitem__(self,pos):

#    if self.n == 0:
#      print('value error----->> list is empty')
    if 0< pos<self.n:

      for i in range(pos,self.n - 1):
        self.A[i] = self.A[i+1]
      self.n = self.n -1

# remove function
  def remove(self, item):
    pos = self.find(item)
    if type(pos) == int:
      self.__delitem__(pos)
```
