We would be using divide and conquer algo to find minimum and maximum value from a array

```python
import sys
```

```python
sys.setrecursionlimit(1500)
```

```python
def findMinMax(arr,i,j):
# small problem
  # for single element
  if i == j:
    min = arr[i]
    max = arr[j]
 # for 2 elements in array
  elif i == j-1:
    if arr[i]<arr[j]:
      min = arr[i]
      max = arr[j]
    else:
      min = arr[j]
      max = arr[i]
# building the divide block
  # big problem
  else:
    mid = int(i +((j-i) // 2))
    ## calling reursion---. conquer
    min_1,max_1 = findMinMax(arr,i,mid)
    min_2,max_2 = findMinMax(arr,mid+1,j)
    # combine stage for divide and conquer
    if min_1 < min_2:
      min = min_1
    else:
      min = min_2
    if max_1 < max_2:
      max = max_2
    else:
      max = max_1
  return min,max



#driver code
arr = [20,39,45,65,21,41,44,89,92]

min_value, max_value = findMinMax(arr,0,8)
print(min_value,max_value)
```

Finding power of an element (Amazom mostly asked questions)

```python
def findPowerofElement(a,n):
  if n == 1:
    return a
  elif n == 0:
    return 1
  else:
    # divide
    mid = n//2
    #conquer
    b = findPowerofElement(a,mid)
    # combine part
    result = b * b
    if n%2 == 0:
      return result
    else:
      return result * 2

# driver code
a = 2
n = 16
result = findPowerofElement(a,n)
print(result)
```

Count number of ways to reach a nth stair
We used here the concept of fibonacci series as number of stairs increases number of ways how to reach follows fibonacci series

```python
def nstairout(n):
  if n == 0:
    return 0
  elif n == 1:
    return 1
  elif n == 2:
    return 2
  else:
    return nstairout(n-1) + nstairout(n-2)

# driver code
nstairout(10)
```
